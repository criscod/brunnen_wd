```python
def changeLetter(row):
    return row[1:]+row[0:1]
```

```python
data['X_dms_a'] = data['X_dms'].apply(changeLetter)
```

```python
data['Y_dms_a'] = data['Y-dms'].apply(changeLetter)
```

```python
datac = data[['X','Y','X_dms_a','Y_dms_a','primaryind','nummer','brunnenart','historisch','wasserart_','bezeichnun']]
```

```python
def generateFakeIds(row):
    return 'QX'+str(row)
```

```python
data_qs=pd.DataFrame()
```

```python
data_qs['qid']=datac['primaryind'].apply(generateFakeIds)

```

```python
def labelquotes(row):
    label = row[9]
    print(label)
    labelr = "\'"+str(label)+"\'"
    return labelr
```

```python
data_qs['a1'] ='Lde'
data_qs['Lde']=datac.apply(labelquotes,axis=1)
data_qs.replace("'nan'","",inplace=True)
#datac['bezeichnun']
```

```python
data_qs['a2'] = 'P31'
data_qs['P31']='Q1630622' #drinking fountain

data_qs['a3'] = 'P31'
data_qs['P31_2']=datac['brunnenart']
# Replacing with the created IDs
#'öffentlicher Brunnen' Q53628296,
#       'Notwasserbrunnen' Q53628522,
#       'privater Brunnen' Q53629707,
#       'Brunnen in städtischer Liegenschaft' Q53628618,
#       'Brunnen des Verschönerungsvereins' Q53628761,
#       'Brunnen mit eigener Versorgung' Q53630002 
data_qs.replace('öffentlicher Brunnen', 'Q53628296', inplace=True)
data_qs.replace('Notwasserbrunnen', 'Q53628522', inplace=True)
data_qs.replace('privater Brunnen', 'Q53629707', inplace=True)
data_qs.replace('Brunnen in städtischer Liegenschaft', 'Q53628618', inplace=True)
data_qs.replace('Brunnen des Verschönerungsvereins', 'Q53628761', inplace=True)
data_qs.replace('Brunnen mit eigener Versorgung', 'Q53630002', inplace=True)
data_qs
```

```python
def processCoordinate(row):
    r = '@'+str(row['Y']) +'/' +str(row['X'])
    return r
```

```python
data_qs['a4'] = 'P625'
data_qs['P625']=datac.apply(processCoordinate,axis=1)
```

```python
def yearformat(row):
    normaly = row[7]
    year='+' + str(normaly)+'-00-00T00:00:00Z/9.'
    print(year)
    return year
```

```python
year = datac.apply(yearformat,axis=1)
```

```python
data_qs['a5'] = 'P571'
data_qs['P571']=year
```

```python
data_qs['a6'] = 'P528' #catalogue ID
data_qs['P528']=datac['nummer']

#qualifier catalogue
data_qs['a7'] = 'P972' #catalogue ID
data_qs['P972']='Q53629101'
```
